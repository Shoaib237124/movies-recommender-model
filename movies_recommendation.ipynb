```python
import numpy as np 
import pandas as pd 
from sklearn.feature_extraction.text import CountVectorizer
```

## Data Preprocessing

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(3)
```

```python
credits.head(3)
```

### Merge Both datasets

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(3)
```

```python
# filter columns which we need
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head(2)
```

```python
# check missing values
movies.isnull().sum()
```

```python
# drop the missing values
movies = movies.dropna()
movies.isnull().sum()
```

```python
# check for duplicates
movies.duplicated().sum()
```

```python
# function to convert string representation of list to actual list
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(3)
```

```python
# define function to extract top 3 elements from a list
def extract_top_3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(extract_top_3)
movies.head(3)
```

```python
# define a function to extract the director's name from the crew
def get_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(get_director)
movies.head(3)
```

```python
# define a function to remove spaces from the list elements
def remove_spaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ", ""))
    return L1
```

```python
movies['genres'] = movies['genres'].apply(remove_spaces)
movies['keywords'] = movies['keywords'].apply(remove_spaces)
movies['cast'] = movies['cast'].apply(remove_spaces)
movies['crew'] = movies['crew'].apply(remove_spaces)

movies.head(3)
```

```python
# convert overview to list of words
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head(3)
```

```python
# created a new column 'tags' which is a combination of all the columns
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head(3)
```

```python
movies.head(3)
import pickle
# create sepaeate pickle file of this dataframe
with open('Combined_Movies.pkl', 'wb') as f:
    pickle.dump(movies, f)
```

```python
# drop other columns
new_movies = movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'])
new_movies.head(3)
```

```python
# convert tags i.e; list to string
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head(3)
```

```python
# steming the tags
from nltk.stem import PorterStemmer
ps = PorterStemmer()

def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)

new_movies['tags'] = new_movies['tags'].apply(stem)
new_movies.head(3)
```

## Vectorization

```python
CV = CountVectorizer(max_features=5000, stop_words='english')
vectors = CV.fit_transform(new_movies['tags']).toarray()
```

```python
vectors.shape
```

```python
# calculate cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(vectors)
```

```python
cosine_sim.shape
```

```python
# define a function to recommend movies based on the title
def recommend_movies(movie):
    matches = new_movies[new_movies['title'].str.lower() == movie.lower()]
    if matches.empty:
        print(f"Movie '{movie}' not found in dataset.")
        return
    
    index = matches.index[0]
    distances = sorted(list(enumerate(cosine_sim[index])), key=lambda x: x[1], reverse=True)
    
    for i in distances[1:6]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend_movies('Terminator 2: Judgment Day')
```

```python
# convert to pickle
import pickle
with open('movies_recommendation.pkl', 'wb') as f:
    pickle.dump(new_movies, f)
    
with open('cosine_similarity.pkl', 'wb') as f:
    pickle.dump(cosine_sim, f)
```
